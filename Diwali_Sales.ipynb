* Total number of orders from top 10 states .

```python
total_orders_per_state = df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False)
total_orders_per_state .head(10)

sea.set(rc={'figure.figsize' : (25,5)}) # in those bracket first one is Width and second one is breadth. 
axis = sea.barplot(x = 'State', y='Orders',data = total_orders_per_state )
for bars in axis.containers :
    axis.bar_label(bars)

```

From above Graph we can see state of highest orders is Utter Pradesh and lowest orders is Telengana . 

* Total amount/sales from top States .


```python
Total_sales_per_state = df.groupby('State',as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending= False)
display(Total_sales_per_state)

Graphical_bars = sea.barplot(data = Total_sales_per_state , x= 'State' , y= 'Amount' )
for bars in Graphical_bars.containers :
    Graphical_bars.bar_label(bars)
```

From the above graph we can that most of the sales are from Utter Pradesh , Maharashtra and Karnataka respectively. 

* Martial Status

##### Product Catagory

* Count total Product catagory.

```python
sea.set(rc={'figure.figsize' : (27,5)})
axis = sea.countplot(data = df , x = 'Product_Category')
for bars in axis.containers:
    axis.bar_label(bars)
```

* Total Sales of each Purchased_catagory 

```python
Total_sales_of_product_catagory = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
display(Total_sales_of_product_catagory)


# Graphical Representation .
sea.set(rc={'figure.figsize' : (35,5)})
Graph = sea.barplot(x='Product_Category' , y='Amount',data = Total_sales_of_product_catagory)
```

* We can from the above graph that highest total sales in category food and clothing category.

Find out total orders in each product category.

```python
Orders_in_product_category = df.groupby('Product_Category',as_index=False)['Orders'].sum()
display(Orders_in_product_category)

# Graphical Representation .
sea.set(rc={'figure.figsize' : (35,10)})
Graph = sea.barplot(x='Product_Category' , y='Orders',data = Orders_in_product_category)
for bars in Graph.containers:
    Graph.bar_label(bars)
```

* From the above graph we can see that people orders more products from Clothing & Apparel and Food category.

Top 10 Highest Orders of Products item.
